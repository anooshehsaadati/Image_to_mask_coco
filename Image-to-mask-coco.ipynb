```python
from pycocotools.coco import COCO
import os
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
file_path="cube segmentation"
```

```python
coco = COCO(file_path + '/valid/_annotations.coco.json')
img_dir = file_path + '/valid'
```

```python
for image_id in range(len(coco.imgs)):
  img = coco.imgs[image_id]
  cat_ids = coco.getCatIds()
  anns_ids = coco.getAnnIds(imgIds=img['id'], catIds=cat_ids, iscrowd=None)
  anns = coco.loadAnns(anns_ids)
  mask = coco.annToMask(anns[0])
  for i in range(len(anns)):
      mask += coco.annToMask(anns[i])

  # for checking unique values in matrix
  res = list(set(i for j in mask for i in j))
  # printing result
  print("Unique values in matrix are : " + str(res))
  mask[mask > 1] = 1
  # for checking unique values in matrix
  res = list(set(i for j in mask for i in j))
  # printing result
  print("Unique values in matrix are : " + str(res))

  # plt.imshow(mask, cmap="gray")
  plt.imsave(file_path + '/valid/mask/mask_'+ img['file_name'], mask, cmap="gray")
```
